```python
import random
```

```python
class Weapons:
    def __init__(self, name, type,rarity="Normal", power=20):
      self.name = name
      self.type = type
      self.rarity = rarity
      self.power = power

    def get_info(self):
      print(f"This is {self.rarity} {self.type}, named {self.name}, it has {self.power} power")

    def attack(self):
      n = random.randint(0,100)
      if n <80:
        print(f"Attack with {self.name} deal {self.power}")
      else:
        print(f"Critical Attact with {self.name} deal 2x damage = {self.power*2}")
```

```python
sword1 = Weapons("Wooden Sword", "Sword")
```

```python
sword1.get_info()
```

```python
sword1.attack()
```

```python
class Skilled_Weapons(Weapons):
  def __init__(self, name, type, rarity="Epic", power=50, skill="Energy Slash"):
    super().__init__(name, type, rarity, power)
    self.__skill = skill
    self.__mana =0

    if self.rarity == "Epic":
      self.__mana_use = 5
    elif self.rarity == "Legendary":
      self.__mana_use = 7

  def attack(self):
    n = random.randint(0,100)
    if n <80:
      print(f"Attack with {self.name} deal {self.power}")
    else:
      print(f"Critical Attact with {self.name} deal 2x damage = {self.power*2}")
    self.__mana += 1

  def get_skill_info(self):
    print(f"This weapon has {self.__skill}, use {self.__mana_use} to activate the skill.")

  def get_mana_info(self):
    print(f"Now this weapon has {self.__mana} mana")

  def use_skill(self):
    if self.__mana < self.__mana_use:
      print(f"Not enough mana to use skill, attack to get more mana")
    else:
      print(f"Use {self.__skill} to deal {self.power*5} damage")
      self.__mana -= self.__mana_use
      print(f"Mana left: {self.__mana}")
```

```python
sword2 = Skilled_Weapons("Magic Sword", "Sword")
```

```python
sword2.get_info()
```

```python
sword2.attack()
```

```python
sword2.get_skill_info()
```

```python
sword2.get_mana_info()
```

```python
sword2.use_skill()
```
