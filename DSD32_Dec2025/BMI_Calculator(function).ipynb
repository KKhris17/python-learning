```python
import time
```

```python
def typing(text, delay=0.02):
  for char in text:
    print(char, end="", flush=True)
    print("|", end="", flush=True)
    time.sleep(delay)
    print("\b \b", end="", flush=True)
    time.sleep(delay)
```

```python
#test function and adjust the delay
typing("Testing typing function")
```

```python
def bmi_calculator(Weight=None,Height=None):
  if (Weight == None):
    typing("Welcome to BMI calculator\n")
    typing("Press Enter to continue")
    input("")
    while (True):
      typing("Please enter your weight (kg)")
      Weight_str = input(":")
      try:
        Weight = float(Weight_str)
        break
      except ValueError:
        typing("Numbers only (kg)\n")
        time.sleep(0.1)
  if (Height == None):
    while (True):
      typing("Please enter your height (cm)")
      Height_str = input(":")
      try:
        Height = float(Height_str)
        break
      except ValueError:
        typing("Numbers only (cm)\n")
        time.sleep(0.1)
  time.sleep(0.2)
  user_bmi = round((Weight/((Height/100)**2)), 2)
  if (user_bmi < 18.5):
    user_bmi_status = "Underweight"
  elif (18.5 <= user_bmi <25):
    user_bmi_status = "Normal"
  elif (25 <= user_bmi <30):
    user_bmi_status = "Overweight"
  elif (30 <= user_bmi <35):
    user_bmi_status = "Obese class 1"
  elif (35 <= user_bmi <40):
    user_bmi_status = "Obese class 2"
  else:
    user_bmi_status = "Obese class 3"
  typing(f"Your BMI : {user_bmi:.2f} {user_bmi_status}")
  return user_bmi,user_bmi_status
```
